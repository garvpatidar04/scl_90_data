```python
def result(model, X, y, X_val, y_val, f):
    X, X_val = X.iloc[:,:f], X_val.iloc[:,:f]
    
    model.fit(X,y)
    
    y_pred = model.predict(X_val)
    
    acc = accuracy_score(y_val, y_pred)
    
#     print(f'Model Accuracy with {f} columns out of 90 is {acc}')
    
    return np.round( (acc*100), 3)
```

```python
result_model = {}
for i in range(1,91):
    result_score = result(
    MultiOutputClassifier(RandomForestClassifier()),
    X,
    y,
    X_val,
    y_val,
    i
    )
    # print(result_score)
    result_model[f'{i}_num_col'] = result_score

    
```

```python
result_log = pd.DataFrame(
    result_model.values(),
    index=result_model.keys(),
    columns = ['accuracy_score']
).reset_index()
```

```python
index = result_log['index']
result_log['col_num'] = index
result_log.drop(columns=['index'], axis=1, inplace=True)

result_log.head()
```
